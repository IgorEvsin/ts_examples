```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
```

```python
# https://www.eia.gov/dnav/pet/pet_cons_wpsup_k_w.htm

data = pd.read_excel(r"../data/oil_consumption.xls", index_col=0)
data.index = pd.to_datetime(data.index)
```

```python
total_consumption = data.iloc[:, 0]  # .iloc[:900, 0]
total_consumption = total_consumption.fillna(method="ffill")
```

```python
plt.plot(total_consumption)
plt.title(f"{data.columns[0]}")
plt.show()
```

```python
stl = STL(total_consumption, seasonal=13)
res = stl.fit()
fig = res.plot()
```
