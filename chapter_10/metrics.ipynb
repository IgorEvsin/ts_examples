```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
wheat = yf.download("ZW=F")
wheat.index = pd.to_datetime(wheat.index)
```

```python
close = wheat[("Close", "ZW=F")]
plt.plot(close, c="0", linewidth="0.5")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

ts_train, ts_test = train_test_split(close, shuffle=False, train_size=0.8)
```

```python
from sktime.forecasting.arima import ARIMA

forecaster = ARIMA()

forecaster.fit(ts_train)
```
