```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
wheat = yf.download("ZW=F")
```

```python
close = wheat[("Close", "ZW=F")]
```

```python
plt.plot(close, c="0", linewidth="0.5")
plt.show()
```

```python
from statsmodels.tsa.seasonal import STL

stl = STL(close, seasonal=13)
res = stl.fit()
fig = res.plot()
```
